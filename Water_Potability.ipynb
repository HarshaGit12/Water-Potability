##Handling missing or null data

```python
def replace_nan(data1 , variable):
    data1[variable] = data1[variable].fillna(data1[variable].mean())

replace_nan(data1 , "ph")
replace_nan(data1 , "Sulfate")
replace_nan(data1 , "Trihalomethanes")
```
